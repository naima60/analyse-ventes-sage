# Importer les librairies

```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Chargement du fichier 

```python
df = pd.read_csv("ventes.csv")
```

# Infos fichier

```python
print(f"Le nombre de ligne et de colonnes : {df.shape}")
print(f"Infos fichiers :")
print(df.info())
print(f"Statestiques générales : \n {df.describe()}")
print(f"Jeux de données des 5 première lignes : \n {df.head()}")
```

# Quel est le chiffre d’affaires total ?

```python
# ajouter colone CA au df
df = df.assign(chiffre_affaires = df['prix'] * df['quantite'])
#print(df.head())
ca_total = df['chiffre_affaires'].sum()
print(f"Le chiffre d'affiare total est {ca_total} €")
```

# Quels sont les produits les plus vendus ?

```python
ca_produits = df.groupby('produit')['chiffre_affaires'].sum()
tri_produits = ca_produits.sort_values()
print(f"le top 3 des produits les plus vendus est : \n {tri_produits.tail(3)}")
```

# Quelle catégorie rapporte le plus ?

```python
# Quelle catégorie rapporte le plus ?
ca_categorie = df.groupby('categorie')['chiffre_affaires'].sum()
top_ca_cat = ca_categorie.sort_values().tail(1)
print(f"La catégorie qui rapporte le plus de CA:\n {top_ca_cat}")

# Répartition CA par catégorie en %
repartition_cat = round((ca_categorie / ca_categorie.sum()) * 100, 2)
print(repartition_cat )
```

```python
# détache la part la plus importante du graphique (la meilleure catégorie)
idx_max = repartition_cat.values.argmax()   # .argmax() → retourne l’indice de la valeur maximale

# Créer la liste explode de la même longueur que le nombre de catégories 
explode = [0.1 if i == idx_max else 0 for i in range(len(repartition_cat))] # 0.1 → la part sera légèrement détachée
                                                                            # 0 → les autres parts restent normales
```

```python
# Visualisation du CA par catégorie en %
plt.pie(
    x= repartition_cat.values,   # affiche la valeur des %
    labels= list(repartition_cat.index),   # affiche les noms de catégories
    autopct = '%.2f%%',   # 2f : nombre de décimal
    textprops={'color':'black', 'fontsize': 10},   # propriété nom catégorie
    startangle=90, 
    shadow=True,     # ajoute l'ombre
     explode=explode  # détache la part la plus importante
    )

# Modifier les % après le tracé pour qu’ils soient blancs
for text in plt.gca().texts:
    if '%' in text.get_text():
        text.set_color('white')         # couleur blanche des %
        text.set_fontweight('bold')     # texte en gras des % 


#plt.title("Répartition CA par catégorie") # par défaut le titre s'affiche en haut
plt.title("Répartition CA par catégorie", fontsize=12, fontweight='bold', y=-0.05)  # afficher le titre en bas

plt.show()
```

# Quelle ville génère le plus de ventes ?

```python
ca_ville = df.groupby('ville')['chiffre_affaires'].sum()
top_ca_ville = ca_ville.sort_values().tail(1)
print(f"La ville qui génère le plus de CA est :\n {top_ca_ville}")
```

```python
# Visualisation Ca par ville

# 1) trier les valeurs de Z-A
ca_ville = ca_ville.sort_values(ascending=False)

# 2) le graphique

# Ajuster la taille de la figure
plt.figure(figsize=(7,5))

# paramètre du graphique
plt.bar(
    x=ca_ville.index,
    height=ca_ville.values,
    color='#66b3ff'    
)

# Afficher les valeurs sur les bars
for i, val in enumerate(ca_ville.values):
    #plt.text(i, val + 10, f'{val:.0f} €', ha='center', fontsize=10) # affiche les valeurs en haut des bars
    plt.text(i, val/2, f'{val:.0f} €', ha='center', va='center', color='white', fontsize=10, fontweight='bold') # affiche les valeurs au milieu

# Ajouter les labels
plt.xlabel("Villes", fontsize=8, fontweight='bold', color='orange')
plt.ylabel("Chiffre d'affaires (€)", fontsize=8, fontweight='bold', color='orange') 

# le titre
plt.title("Chiffre d'affaires par ville", fontsize=13, fontweight='bold', y=-0.20)

# Attirer l'oeil sur la plus grande barre
colors = ['#ff9999'] + ['#66b3ff']*(len(ca_ville)-1)
plt.bar(ca_ville.index, ca_ville.values, color=colors)

# Afficher le graphique
plt.show()
```

# Mini tableau de bord (combine des deux graphique

```python
# --- Création de la figure ---
fig, axes = plt.subplots(1,2, figsize=(14,6))

# 1️⃣ Camembert CA par catégorie
axes[0].pie( 
    x= repartition_cat.values,   # affiche la valeur des %
    labels= list(repartition_cat.index),   # affiche les noms de catégories
    autopct = '%.2f%%',   # 2f : nombre de décimal
    textprops={'color':'black', 'fontsize': 10},   # propriété nom catégorie
    startangle=90, 
    shadow=True,     # ajoute l'ombre
     explode=explode  # détache la part la plus importante
    )

# Modifier les % après le tracé pour qu’ils soient blancs
for text in plt.gca().texts:
    if '%' in text.get_text():
        text.set_color('white')         # couleur blanche des %
        text.set_fontweight('bold')     # texte en gras des % 

axes[0].set_title("Répartition CA par catégorie", fontsize=13, fontweight='bold', y=-0.05)

# 2️⃣ Bar chart CA par ville
axes[1].bar(x=ca_ville.index,
    height=ca_ville.values,
    color='#66b3ff'    )

# Afficher les valeurs sur les bars
for i, val in enumerate(ca_ville.values):
    #plt.text(i, val + 10, f'{val:.0f} €', ha='center', fontsize=10) # affiche les valeurs en haut des bars
    plt.text(i, val/2, f'{val:.0f} €', ha='center', va='center', color='white', fontsize=10, fontweight='bold') # affiche les valeurs au milieu

# Ajouter un titre
axes[1].set_title("Chiffre d'affaires par ville", fontsize=13, fontweight='bold', y=-0.20)

# Ajouter les labels
axes[1].set_xlabel("Villes", fontsize=8, fontweight='bold', color='orange')
axes[1].set_ylabel("Chiffre d'affaires (€)", fontsize=8, fontweight='bold', color='orange') 

# Attirer l'oeil sur la plus grande barre
colors = ['#ff9999'] + ['#66b3ff']*(len(ca_ville)-1)
plt.bar(ca_ville.index, ca_ville.values, color=colors)

# --- Ajustement global ---
plt.tight_layout()
plt.show()
```

# Générer une analyse automatiquement

```python
# Récupérer les infos clé
# Catégorie la plus rentable
top_categorie = ca_categorie.idxmax()
ca_top_categorie = ca_categorie.max()

# Ville la plus performante
top_ville = ca_ville.idxmax()
ca_top_ville = ca_ville.max()

# CA total
ca_total = df['chiffre_affaires'].sum()


analyse_texte = f"""
ANALYSE DES VENTES

- Le chiffre d'affaires total s'élève à {ca_total:,.0f} €.

- La catégorie la plus contributrice au chiffre d'affaires est
'{top_categorie}' avec un CA de {ca_top_categorie:,.0f} €, ce qui représente {(ca_top_categorie / ca_total) * 100:.2f}% du CA total.

- D'un point de vue géographique, la ville de '{top_ville}'
est la plus performante avec un chiffre d'affaires de {ca_top_ville:,.0f} €.
"""

print(analyse_texte)
```

```python
recommandations = f"""
RECOMMANDATIONS

- Consolider la gamme '{top_categorie}', qui représente une part significative du chiffre d'affaires.

- Renforcer les actions commerciales dans la ville de '{top_ville}', zone actuellement la plus rentable.

- Analyser les catégories à faible CA afin d’identifier des leviers d’optimisation (prix, volume, abandon).
"""

print(recommandations)
```
